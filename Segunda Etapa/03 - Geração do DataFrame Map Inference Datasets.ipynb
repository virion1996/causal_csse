```python
import zipfile
import os

# Definir o nome do arquivo zip
zip_filename = "causal_csse.zip"

# Lista dos arquivos que deseja zipar
arquivos_para_zipar = ["csse.py", "prepare_dataset.py", "01 - Geração de métricas para todas as bases - periódico.ipynb"]

# Criar o arquivo zip e adicionar os arquivos
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for arquivo in arquivos_para_zipar:
        zipf.write(arquivo, os.path.basename(arquivo))

print(f"Arquivos compactados com sucesso em {zip_filename}")
```
