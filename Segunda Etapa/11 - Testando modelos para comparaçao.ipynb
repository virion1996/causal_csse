```python
from carla import OnlineCatalog, MLModelCatalog
from carla.recourse_methods import CRUDS

dataset = OnlineCatalog("adult")      # exemplo com dataset "adult"
model = MLModelCatalog(dataset, "ann")  # ou outro modelo no catálogo
crud = CRUDS(model)                   # inicializa o método CRUDS
factuals = dataset.raw.iloc[:10]
counterfactuals = crud.get_counterfactuals(factuals)
```
