```python
import os
import boto3
import json
import sagemaker
from datetime import datetime
from sagemaker.processing import ProcessingOutput, ProcessingInput
from sagemaker.network import NetworkConfig
import datetime
from sagemaker.sklearn.processing import SKLearnProcessor


s3 = boto3.client('s3', region_name='us-east-1')
job_name= ""

sklearn_processor = SKLearnProcessor(
    framework_version='1.0-1',  # Escolha a versão do scikit-learn
    role=sagemaker.get_execution_role(),
    instance_type='ml.m5.2xlarge',  # Escolha o tipo de instância
    instance_count=1,  # Quantidade de instâncias
    volume_size_in_gb=50
)

def run(job_name, list_dataset_name, bb_model_name):
    sklearn_processor.run(
         code= 'pj_causal_csse_metricas.py',#os.environ.get("ScriptPath"),
         wait=False,
         job_name=job_name,
         arguments=[
             "--list_dataset_name", list_dataset_name,
             "--bb_model_name", bb_model_name,
         ],
         inputs=None,
         outputs= None
     )
```
